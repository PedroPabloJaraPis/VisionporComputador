```python
import cv2
import numpy as np

# Cargar la imagen
im = cv2.imread('/content/drive/MyDrive/Vision por computador/Semana 4/cat.jpg')

# Definir el color de fondo deseado (en formato BGR)
color_fondo = (255, 0, 0 )  # Rojo

# Crear una mscara para el objeto principal (suponiendo que el fondo es blanco)
mascara_objeto = np.all(im != [0, 255, 0], axis=-1)

# Crear una imagen con el nuevo fondo rojo
nuevo_fondo = np.full_like(im, color_fondo, dtype=np.uint8)

# Copiar el objeto principal de la imagen original al nuevo fondo
imagen_resultante = np.where(mascara_objeto[:, :, None], im,nuevo_fondo)

# Mostrar la imagen resultante
imagen_resultante
```

```python
color_fondo = (0, 0, 255)  # Rojo

# Crear una mscara para el objeto principal (suponiendo que el fondo es blanco)
mascara_objeto = np.all(im != [255, 255, 255], axis=-1)

# Crear una imagen con el nuevo fondo rojo
nuevo_fondo = np.full_like(im, color_fondo, dtype=np.uint8)

# Copiar el objeto principal de la imagen original al nuevo fondo
imagen_resultante = np.where(mascara_objeto[:, :, None], im, nuevo_fondo)

# Mostrar la imagen resultante
imagen_resultante
```

```python
color_fondo = (0, 0, 255)  # Rojo

# Definir el umbral de color para el fondo verde
umbral_bajo = np.array([0, 100, 0], dtype=np.uint8)
umbral_alto = np.array([100, 255, 100], dtype=np.uint8)

# Crear una mscara para identificar el fondo verde
mascara_fondo_verde = cv2.inRange(im, umbral_bajo, umbral_alto)

# Invertir la mscara para obtener el objeto principal
mascara_objeto = cv2.bitwise_not(mascara_fondo_verde)

# Crear una imagen con el nuevo fondo rojo
nuevo_fondo = np.full_like(im, color_fondo, dtype=np.uint8)

# Copiar el objeto principal de la imagen original al nuevo fondo
imagen_resultante = cv2.bitwise_and(im, im, mask=mascara_objeto)
imagen_resultante += cv2.bitwise_and(nuevo_fondo, nuevo_fondo, mask=mascara_fondo_verde)

# Mostrar la imagen resultante
imagen_resultante
```

```python
def disolucion():
    # Cargar las dos imgenes que quieres disolver
    imagen_0 = cv.imread('CiudadD.png')
    imagen_1 = cv.imread('CiudadN.png')

    imagen_1 = cv.resize(imagen_1, (imagen_0.shape[1], imagen_0.shape[0]))

    # Definir el parmetro de disolucin (de 0 a 1)
    a = 0.5  # Por ejemplo, una disolucin al 50%

    # Aplicar la frmula de disolucin
    disolucion = cv.addWeighted(imagen_0, 1-a, imagen_1, a, 0)

    # Mostrar la imagen resultante de la disolucin
    cv.imshow('Imagen de Disolucin', disolucion)
    cv.waitKey(0)
    cv.destroyAllWindows()
```
