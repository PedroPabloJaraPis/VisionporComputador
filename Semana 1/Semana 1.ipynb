```python
h, w = 300, 300
image = np.zeros((h, w, 3), np.uint8)

row = (h - 80) // 2
col = (w - 80) // 2

for i in range (80):
  for j in range (80):
    image[row + i, col + j] = (0, 0, 255)

row2 = (h - 60) // 2
col2 = (w - 60) // 2
for i in range (60):
  for j in range (60):
    image[row2 + i, col2 + j] = (0, 255, 0)

row3 = (h - 40) // 2
col3 = (w - 40) // 2
for i in range (40):
  for j in range (40):
    image[row3 + i, col3 + j] = (255, 0, 0)

cv2_imshow(image)
```
